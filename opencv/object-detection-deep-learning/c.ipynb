```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('images/example_01.jpg')
img
```

```python
type(img)
```

```python
plt.imshow(img)
```

```python
normal=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(normal)
```

```python
def get_image(image,cmap=None):
    plt.axis('off')
    plt.imshow(image,cmap=cmap)
    plt.show()
    
```

```python
get_image(normal)
```

```python
hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
get_image(hsv)
```

```python
type(hsv)
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
get_image(gray,cmap='gray')
```

# Bllurred

```python
width,height= normal.shape[:2]
print(width)
print(height)

if(width>500):
    k=(50,50)
elif(width>200) and (width<=500):
    k=(25,25)
else:
    k=(10,10)
blur=cv2.blur(normal,k)
get_image(blur)
```

```python
v,binary=cv2.threshold(gray,120,255,cv2.THRESH_BINARY)
get_image(binary,cmap='gray')
```

```python
invert=cv2.bitwise_not(binary)
get_image(invert, cmap='gray')
type(invert)
```
